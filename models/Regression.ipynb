```python
import pandas as pd

# Load the dataset
file_path = '../../data/processed/merged_cycling_data.csv'
df = pd.read_csv(file_path)

#Convert ADDRESSNUMBERTEXT to string
df['ADDRESSNUMBERTEXT'] = df['ADDRESSNUMBERTEXT'].astype(str)

# Convert datetime columns to datetime type
df['datetime_bin_start'] = pd.to_datetime(df['datetime_bin_start'])
df['datetime_bin_end'] = pd.to_datetime(df['datetime_bin_end'])


# Feature selection ( dropping irrelevant columns including 'geometry')
irrelevant_features = ['_id', 'ID', 'geometry', 'index_right', 'SDE_STATE_ID','centreline_id']
df.drop(columns=irrelevant_features, inplace=True)

# Display the first few rows of the dataset
print(df.head())
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, LabelEncoder

#  Normalize/Scale features
scaler = StandardScaler()
numerical_features = ['daily_temperature', 'daily_precipitation', 'distance_meters']
df[numerical_features] = scaler.fit_transform(df[numerical_features])

#  Encode categorical variables
label_encoder = LabelEncoder()
categorical_features = ['ADDRESSNUMBERTEXT','ADDRESSSTREET', 'FROMSTREET', 'DIRECTION', 'WARD', 'BIA', 'ASSETTYPE', 'STATUS', 'direction', 'location', 'class_type']
for col in categorical_features:
    df[col] = label_encoder.fit_transform(df[col].astype(str))



# Now df is cleaned and ready for model training
print(df.head())
```

```python
df.describe()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Assuming 'df' is your DataFrame
# df = pd.read_csv('your_data.csv')  # Uncomment and modify this line to load your data

# Convert datetime features to numerical features
for column in df.select_dtypes(include=['datetime64']).columns:
    df[column + '_year'] = df[column].dt.year
    df[column + '_month'] = df[column].dt.month
    df[column + '_day'] = df[column].dt.day
    df[column + '_hour'] = df[column].dt.hour
    df[column + '_minute'] = df[column].dt.minute
    df.drop(columns=[column], inplace=True)


# Fill missing values with the mean of the respective columns
df.fillna(df.mean(), inplace=True)


# Assuming 'volume' is the target variable
target = 'volume'
features = df.drop(columns=[target])
target_values = df[target]

# Convert categorical variables to dummy/indicator variables
features = pd.get_dummies(features)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target_values, test_size=0.3, random_state=42)

# Initialize the model
model = LinearRegression()

# Train the model
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error: {mse}')
print(f'R^2 Score: {r2}')
```
